````python
def get_response(user_content, system_content, model):
    completion =  openai.ChatCompletion.create(
    model= model, #"gpt-4-0613",#"gpt-3.5-turbo-0613", gpt-4o
    temperature=0,
    messages=[
    {"role": "system", "content": system_content},
    {"role": "user", "content": user_content}
  ]
)
    return(completion.choices[0].message.content)
````

````python
import json
def responses_dictionary(df,
                        source_language,
                        target_language,
                        original_text_row,
                        translation_Google_row,
                        translation_User_row ):
  dict_responses = []
  df_results = pd.DataFrame()
  for index, row in df[0:].iterrows():
    original_text = row[original_text_row] # row['English Original']
    translation_Google = row[translation_Google_row] # row['Finnish MT (Google Translate)']
    translation_User = row[translation_User_row]
    instruction = translation_evaluation(source_language, target_language, original_text, translation_Google, translation_User)

    dict_response = get_response(instruction, system_prompt, 'gpt-4o')
    dict_responses.append(dict_response.replace('```json','').replace('```',''))
    try:
      dicts = [json.loads(d) for d in dict_responses]
      df_results = pd.DataFrame(dicts)
    except:
      df_results = dict_responses

  return df_results

# Example of use
# responses_dictionary('English','Finnish', 'English Original', 'Finnish MT (Google Translate)', 'Finnish Human Reference')

````
